In this notebook we explore the ADNI1 mask, and see how to move it into the same region as the ADNI2 data we've created. 

This notebook was used to create `04_mask/mask.py`. 

```python
import tables as tb
import numpy as np
import pyminc.volumes.factory as mnc
%matplotlib inline
```

```python
fileh = tb.open_file('/projects/nikhil/ADNI_prediction/adni-autoencoder/combined.h5', mode='r', title='data')
l_datamask = fileh.root.l_datamask[:]
r_datamask = fileh.root.r_datamask[:]
l_vol = np.zeros(fileh.root.l_volmask.shape)
```

### What's the difference betwen volmask and datamask?

```python
f = figure()
axis = 2

# volmask
p = f.add_subplot(1, 2, 1)
imshow(np.sum(fileh.root.l_volmask, axis=axis))
p.set_title('volmask')

# datamask
p = f.add_subplot(1,2,2)
vol = np.zeros(fileh.root.l_volmask.shape)
flatvol = vol.reshape(-1)
flatvol[l_datamask] = 1
vol = flatvol.reshape(fileh.root.l_volmask.shape)
imshow(np.sum(vol, axis=axis))
p.set_title('datamask')
```

### Move the datamask from ADNI1 to the cropbox of ADNI2 images

```python
f = figure(figsize=(15,20))

# step 1: reconstitute the 3D datamask
p = f.add_subplot(1,3,1)
adni1_l_datamask = fileh.root.l_datamask[:]
adni1_l_datavol = np.zeros(fileh.root.l_volmask.shape).reshape(-1)
adni1_l_datavol[adni1_l_datamask] = 1
adni1_l_datavol = adni1_l_datavol.reshape(fileh.root.l_volmask.shape)
imshow(np.sum(adni1_l_datavol, axis=axis))
p.set_title('datamask in ADNI1 cropbox')

# step 2: insert the 3D datamask into the whole volume
wholevol = np.zeros(fileh.root.l_fullvolmask.shape)
cmin = fileh.root.l_cropbbox_min
cmax = fileh.root.l_cropbbox_max
wholevol[cmin[0]:cmax[0],cmin[1]:cmax[1],cmin[2]:cmax[2]] = adni1_l_datavol


p = f.add_subplot(1,3,2)
imshow(np.sum(wholevol, axis=axis))
p.set_title('ADNI1 whole volume datamask')

# step 3: extract the datamask using the ADNI2 cropbbox
adni2h = tb.open_file('03_clean/candidate_left.h5', mode='r', title='data')
cmin = adni2h.root.cropbbox_min
cmax = adni2h.root.cropbbox_max
adni2_l_datavol = wholevol[cmin[0]:cmax[0],cmin[1]:cmax[1],cmin[2]:cmax[2]]
p = f.add_subplot(1,3,3)
imshow(np.sum(adni2_l_datavol, axis=axis))
p.set_title('datamask in ADNI2 cropbox')
```

```python
# what does the ADNI2 mask look like
imshow(np.sum(adni2h.root.mask, axis=axis))
```

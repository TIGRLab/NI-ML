This is a test of the masking of the ADNI2 data using the ADNI1 mask. 
In this test, we load up the masked ADNI2 data, and reconstitute the 3D volume according to the mask contained in the HDF5 file.
Hopefully it looks good!

```python
import tables as tb
import numpy as np
%matplotlib inline
fileh = tb.open_file('04_mask/fused_right.h5', mode='r', title='data')
datamask = fileh.root.datamask[:]
datamaskflat = datamask.reshape(-1)

data = fileh.root.data[:]
datarecon = np.zeros(shape=[data.shape[0]] + list(datamask.shape))
datareconflat = datarecon.reshape(data.shape[0],-1)
datareconflat[:,datamaskflat] = data
datarecon = datareconflat.reshape(datarecon.shape)
imshow(np.sum(np.sum(datarecon,axis=0), axis=0))
```
